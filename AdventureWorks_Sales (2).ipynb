```python
# Convert the column to datetime format
df['ShipDateKey'] = pd.to_datetime(df['ShipDateKey'], format='%Y%m%d')

# Convert the datetime column to the desired string format
df['ShipDateKey'] = df['ShipDateKey'].dt.strftime('%Y-%m-%d')

# Print the DataFrame to see the conversion
print(df)
```

```python
# Convert the column to datetime format
df['DueDateKey'] = pd.to_datetime(df['DueDateKey'], format='%Y%m%d')

# Convert the datetime column to the desired string format
df['DueDateKey'] = df['DueDateKey'].dt.strftime('%Y-%m-%d')
```

```python
# Convert the column to datetime format
df['OrderDateKey'] = pd.to_datetime(df['OrderDateKey'], format='%Y%m%d')

# Convert the datetime column to the desired string format
df['OrderDateKey'] = df['OrderDateKey'].dt.strftime('%Y-%m-%d')
```
